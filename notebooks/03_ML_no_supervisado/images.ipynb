```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generar datos de ejemplo
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Aplicar K-means
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Graficar
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("K-means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate synthetic data for health (patients) and logistics (delivery points)
np.random.seed(42)

# Health data: [Age, Disease Severity, Treatment Cost, Length of Stay]
health_data = np.array([
    [np.random.randint(20, 80), np.random.uniform(1, 10), np.random.randint(1000, 20000), np.random.randint(1, 30)]
    for _ in range(50)
])

# Logistics data: [Latitude, Longitude, Delivery Volume, Delivery Frequency]
logistics_data = np.array([
    [np.random.uniform(-5, 5), np.random.uniform(-5, 5), np.random.randint(10, 500), np.random.randint(1, 20)]
    for _ in range(50)
])

# Combine data and create DataFrames for clarity
health_df = pd.DataFrame(health_data, columns=["Age", "Severity", "Cost", "LengthOfStay"])
logistics_df = pd.DataFrame(logistics_data, columns=["Latitude", "Longitude", "Volume", "Frequency"])

# Perform hierarchical clustering
# Using 'ward' linkage for compact clusters
health_linkage = linkage(health_data, method='ward')
logistics_linkage = linkage(logistics_data, method='ward')

# Plot dendrogram for health data
plt.figure(figsize=(10, 6))
plt.title("Dendrogram for Health Data (Patients)")
plt.xlabel("Patient Index")
plt.ylabel("Distance")
dendrogram(health_linkage, leaf_rotation=90., leaf_font_size=8.)
plt.show()

# Plot dendrogram for logistics data
plt.figure(figsize=(10, 6))
plt.title("Dendrogram for Logistics Data (Delivery Points)")
plt.xlabel("Delivery Point Index")
plt.ylabel("Distance")
dendrogram(logistics_linkage, leaf_rotation=90., leaf_font_size=8.)
plt.show()
```
